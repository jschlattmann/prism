```sql
--CREATE OR REPLACE TABLE teamconnect.pca_prism.09_tbl_prm_active AS
INSERT OVERWRITE TABLE teamconnect.pca_prism.09_tbl_prm_active
SELECT
  btn,
  dpi_environment,
  COLLECT_LIST(
    named_struct(
      'promo_code',  promo_code,
      'promo_start', promo_start,
      'promo_end',   promo_end,
      'promo_amount', promo_amount
    )
  ) AS active_promos
FROM teamconnect.pca_prism.08_tbl_prm_shaped
WHERE (promo_start IS NULL OR promo_start <= current_date())
  AND (promo_end   IS NULL OR promo_end   >= current_date())
GROUP BY btn, dpi_environment;


OPTIMIZE teamconnect.pca_prism.09_tbl_prm_active
ZORDER BY (btn, dpi_environment);
```
